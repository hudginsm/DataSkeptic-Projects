```python
import pandas as pd
PathName = r'/home/dataskeptic/git_repo/DataSkeptic-Projects/DataSets/Comparing_Hospitals/'
files_of_interest = [
    'hvbp_tps.csv',
    'hvbp_clinical_care.csv',
    'hvbp_safety.csv',
    'hvbp_efficiency.csv',
    'hvbp_hcahps.csv'
]
dfs = {
    foi: pd.read_csv(PathName + foi, header=0) for foi in files_of_interest
}
```

```python
for v in dfs.values():
    for column in v.columns:
        print(column)
    print('\n')
```
