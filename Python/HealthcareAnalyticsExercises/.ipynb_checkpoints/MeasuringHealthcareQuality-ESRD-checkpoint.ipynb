```python
df_states = df.groupby('State').size()
print(df_states)
```

```python
df_states = df.groupby('State').size().sort_values(ascending=False)
print(df_states)
```

```python
df_states = df.groupby('State').size().sort_values(ascending=False).head(n=10)
print(df_states)
```

```python
df_ca = df.loc[df['State'] == 'CA']
print(df_ca)
```

```python
print(df.groupby('Total Performance Score').size())
```

```python
df_filt = df.loc[df['Total Performance Score'] != 'No Score']
df_filt['Total Performance Score'] = pd.to_numeric(df_filt['Total Performance Score'])

df_tps =df_filt[['Facility Name', 'State', 'Total Performance Score']].sort_values('Total Performance Score')
print(df_tps.head(5))
```

```python
import numpy as np
df_state_mean = df_filt.groupby('State').agg({'Total Performance Score': np.mean})
print(df_state_mean.sort_values('Total Performance Score', ascending=False))
```

```python
import numpy as np
df_state_mean = df_filt.groupby('State').agg({
    'Total Performance Score': np.mean,
    'State': np.size
                                             })
print(df_state_mean.sort_values('Total Performance Score', ascending=False))
```
